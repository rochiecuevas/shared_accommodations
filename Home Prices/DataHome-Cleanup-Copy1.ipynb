```python
# Create a path to the csv and read it into a Pandas DataFrame
#csv_path = "Assessor_Historical_Secured_Property_Tax_Rolls.csv"
csv_path = "combine_updated2.csv"
rent_path = "rent_year_csv.csv"
home_df = pd.read_csv(csv_path, encoding='utf8', engine='python')
rent_df = pd.read_csv(rent_path, encoding='utf8', engine='python')

home_df.head(20)
```

```python
rent_df.columns
```

```python
my_col_names = list(rent_df.columns.values)[3:]
melted_df = pd.melt(rent_df,id_vars=["Region Name","Region Type","Data Type"],value_vars=my_col_names,var_name="year",value_name="dollars")
```

```python
rent_df = melted_df.rename(columns={'Region Name': 'Assessor Neighborhood'})
```

```python
home_df.head()
rent_df["year"] = rent_df["year"].astype("int64")
```

```python
home_df = home_df.drop("Unnamed: 0",axis = 1)
```

```python
home_df
```

```python
merged = pd.merge(home_df,rent_df, 
                  how='inner',
                  left_on=["Assessor Neighborhood","Closed Roll Year"],
                  right_on=["Assessor Neighborhood","year"])
merged
```

```python
merged = merged.drop("year", axis = 1)
```
