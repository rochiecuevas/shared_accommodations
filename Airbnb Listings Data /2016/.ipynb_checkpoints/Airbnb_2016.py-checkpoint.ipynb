```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
#csv_path = "20160202 listings.csv"

airbnb20160202_df = pd.read_csv("20160202 listings.csv")

airbnb20160202_df.head()
```

```python
#Create DataFrame with column: "neighbourhood_cleansed","property_type","price"
airbnb20160202_df = airbnb20160202_df[["neighbourhood_cleansed","property_type","price"]]
airbnb20160202_df.head()
```

```python
#Calculate the annual price
daily_data = airbnb20160202_df["price"].str.replace("$","")
daily_data = daily_data.str.replace(",","")
daily_data = daily_data.astype(float)
airbnb20160202_df["annual rate"] = daily_data*365
airbnb20160202_df.head()
```

```python
#Add date to the DataFrame 
airbnb20160202_df["date"] = pd.to_datetime("2016-02")
airbnb20160202_df.head()
```
